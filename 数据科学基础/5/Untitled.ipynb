```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 10, 0.01)
y = (1.5 + np.sin(x)) * (x + np.power(x, 2))
plt.plot(x, y)
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
def inte(n):
    x0 = np.random.uniform(1,10, n)
    y0 = np.random.uniform(0, 200, n)
    result = (1.5 + np.sin(x0)) * (x0 + np.power(x0, 2)) > y0
    return result.sum()/n * 10*200
inte(100000000)
```

```python
nba = pd.read_csv("/home/ran/Downloads/NBA_player_of_the_week.csv")
nba.head()
```

```python
sampleMeanList = []
for i in range(5000):
    sampleMeanList = np.append(sampleMeanList, nba.sample(100).Age.mean())
```

```python
plt.scatter(np.arange(1, 5001), sampleMeanList, alpha = 0.05)
plt.xlabel('index of samples')
plt.ylabel('mean sample age')
plt.plot([0, 5000], [nba.Age.mean(), nba.Age.mean()], color = 'r', lw = 3)
plt.show()
```

```python
plt.hist(sampleMeanList, 30, density=True, color = 'y', edgecolor = 'b')
plt.plot([nba.Age.mean(), nba.Age.mean()], [0,1.3],color = 'r', lw = 3 )
plt.xlabel('mean of sample age')
plt.ylabel('density')
plt.grid(True)
```
