```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2,3,4,5], np.arange(2,20,4))
plt.xlabel("a")
plt.ylabel("b")
plt.title("title")
plt.show()
```

```python
plt.plot(np.arange(1,6),np.arange(2,20,4),'r+--',linewidth = 1,markersize=20)
plt.xticks([1,2,3,4,5])
plt.yticks(np.linspace(0,20,5,endpoint = True))
plt.show()
```

```python
plt.figure(figsize=(8,4),dpi = 90,)#dip分辨率,默认figsize=(4,3)
plt.subplot(2,1,1)#划分制图区域，subplot(row,column,number)
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
c,s = np.cos(x), np.sin(x)
plt.axis([-2*np.pi,2*np.pi,-1,1])
plt.xticks(np.linspace(-np.pi,np.pi,5),[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])#r->row
plt.plot(x,c,'y--',label = 'cos')
plt.plot(x,s,'r',label='sin')
plt.legend(loc = "lower left")


plt.savefig("fff.png",dip=100)
```
