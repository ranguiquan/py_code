```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
family = pd.read_csv("/home/ran/Downloads/GaltonFamilies (1).csv")
family['height'] = family.childHeight * 2.54
family.quantile(0.5)
```

```python
thisbins = np.arange(140,201,10)
n, bins, patches = plt.hist(family.height, thisbins, edgecolor = 'r', color = 'g', density=True)
plt.grid()
plt.xlabel('xxx')
plt.show()
print(bins)
print(n)
print(patches[0])
```

```python
plt.hist(family.sample(10000, replace=True).height, thisbins)
```

```python
mid = []
for i in range(5000):
    mid = np.append(mid, family.sample(100).height.median())
```

```python
plt.scatter(np.arange(1,5001), mid, alpha = 0.01)
plt.plot([0,5000], [family.height.median(),family.height.median()], alpha = 0.5)
plt.plot([0,5000], [mid.mean(), mid.mean()], alpha = 0.5)
plt.ylim(160,180)
plt.show()
```

```python
plt.hist(mid, density=True)
plt.show()
```

```python
nums = pd.DataFrame({'bianhao':np.arange(1,1001)})
```

```python
ans = []
for i in range(10000):
    ans = np.append(ans,max(nums.sample(10, replace=True).bianhao))
plt.scatter(np.arange(1, 10001), ans, alpha = 0.05)
plt.ylim(0,1200)
plt.plot([0,10000], [ans.max(), ans.max()], color = 'r')
print(ans.max())
```

```python
ansmid = []
for i in range(10000):
    ansmid = np.append(ansmid, nums.sample(30, replace=True).bianhao.mean())
```

```python
plt.scatter(np.arange(1, 10001), ansmid, alpha = 0.05)
```

```python
plt.hist(2*ansmid, 100, density = True)
plt.show()
```
