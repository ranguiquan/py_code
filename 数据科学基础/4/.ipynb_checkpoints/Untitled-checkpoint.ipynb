```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, color = 'b', lw = 2, ls = '-', label = "cos")
plt.plot(x, s, color = 'r', lw = 2, ls = '-', label = "sin")
plt.legend(loc = "upper left")
t = 2*np.pi/3
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi], [r'$-\pi$', r'$-\pi/2$', 0, r'$\pi/2$', r'$\pi$'])
plt.yticks([-1,0,1])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data', 0))

plt.annotate(r'$\sin\left(\frac{2\pi}{3}\right)=\frac{\sqrt{3}}{2}$',
            xy = (t, np.sin(t)), xytext = (1.7,1.2), fontsize = 16,
            arrowprops = dict(arrowstyle = '->'))
plt.annotate(r'$\cos\left(\frac{2\pi}{3}\right)=-\frac{1}{2}$', 
            xy = (t, np.cos(t)), xytext = (0.2, -1), fontsize = 16, 
            arrowprops = dict(arrowstyle = '->'))
plt.plot([t,t], [0,np.sin(t)], color = 'r', ls = '--', lw = 2)
plt.plot([t,t], [0,np.cos(t)], color = 'b', ls = '--', lw = 2)

plt.scatter([t,],[np.sin(t),], 50, color='r')
plt.scatter([t,],[np.cos(t),], 50, color='b')
```

```python
a = ['水星','金星','地球','火星','木星','土星','天王星','海王星']
b = [0.241,0.615,1.00,1.88,11.8,29.5,84.0,165]
c = [0.39,0.72,1.00,1.52,5.20,9.54,19.18,30.06]
table = pd.DataFrame({'行星': a, '周期（年）': b, '平均距离': c},
                     columns = ['行星','周期（年）','平均距离'])
table
```

```python
plt.figure(1,figsize=(9,4))
plt.subplot(121)
plt.plot(table.iloc[:,1], table.iloc[:,2])
plt.subplot(122)
plt.plot(table.iloc[:,1]**2, table.iloc[:,2]**3, color = 'b', lw = 5, alpha=0.2)
plt.plot([0,25000],[0,25000],color = 'r', ls='--', alpha = 0.8)
plt.show()
```

```python
studios = np.array(['Warner Bros','Buena Vista (Disney)','Fox', 'Paramount','Universal','Disney',
                                'Columbia','MGM','UA','Sony','New Line','Paramount/Dreamworks','RKO',
                                'Lionsgate','Dreamworks','Tris','Sum','Waner Bros. (New Line)', 'Selz',
                                'Orion','NM','MPC','IFC','AVCO'])

counts = np.array([29,29,26,25,22,11,10,7,6,6,5,4,3,3,3,2,2,1,1,1,1,1,1,1])
movies_and_studios = pd.DataFrame({'Studio':studios,'Count':counts})
movies_and_studios.sort_values(by = 'Count', inplace = True, ascending = False)
movies_and_studios.head()
```

```python
plt.figure(figsize = (9,6))
x_pos = np.arange(len(movies_and_studios.index), 0, -1)
plt.barh(x_pos, movies_and_studios.Count, color = 'y', edgecolor = 'r')
plt.yticks(x_pos, movies_and_studios.Studio)
plt.xlabel('Counts')
plt.grid(True)
plt.show()
```
