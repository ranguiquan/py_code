```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2,3,4,5], np.arange(2,20,4))
plt.xlabel("a")
plt.ylabel("b")
plt.title("title")
plt.show()
```

```python
plt.plot(np.arange(1,6),np.arange(2,20,4),'r+--',linewidth = 1,markersize=20)
plt.xticks([1,2,3,4,5])
plt.yticks(np.linspace(0,20,5,endpoint = True))
plt.show()
```

```python
plt.figure(figsize=(8,4),dpi = 90,)#dip分辨率,默认figsize=(4,3)
plt.subplot(2,1,1)#划分制图区域，subplot(row,column,number)
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
c,s = np.cos(x), np.sin(x)
plt.axis([-2*np.pi,2*np.pi,-1,1])
plt.xticks(np.linspace(-np.pi,np.pi,5),[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])#r->row
plt.plot(x,c,'y--',label = 'cos')
plt.plot(x,s,'r',label='sin')
plt.legend(loc = "lower left")


plt.savefig("fff.png",dip=100)
```

```python
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)
n, bins, patches = plt.hist(x, 100, density = 1, facecolor = 'y',alpha = 0.5)
#alpha 透明度
plt.grid(True)
plt.title('histogram of iq')
plt.xlabel('smart')
plt.ylabel('prob')
plt.text(60,0.025,r'$\mu = 100,\ \sigma = 15$')
plt.show()
```

```python
t = np.arange(0,20,0.01)
s = np.cos(t)
line = plt.plot(t,s,lw = 2)
plt.annotate('local max', xy = (0,1), xytext = (2,1.5), arrowprops = dict(facecolor = 'b', shrink = 0.05))
#shrink 让箭头缩小
plt.ylim(-2,2)
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, color = 'b', lw = 2, ls = '--', label = "cos")
plt.plot(x, s, color = 'r', lw = 2, ls = '-', label = "sin")
plt.legend(loc = "upper left")
t = 2*np.pi/3
plt.annotate(r'$\sin\left(\frac{2\pi}{3}\right)=\frac{\sqrt{3}}{2}$',
            xy = (t, np.sin(t)), xytext = (1.7,1.2), fontsize = 16,
            arrowprops = dict(arrowstyle = '->'))
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.xticks(np.arange(-np.pi, np.pi, np.pi/2))
plt.scatter([t,],[np.sin(t),], 50, color='r')
plt.scatter([t,],[np.cos(t),], 50, color='b')
plt.plot([t,t], [0,np.cos(t)], color = 'b', ls='--',lw=2)
```
